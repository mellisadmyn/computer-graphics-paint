```python
# Python implementation of the Bresenham Line Algorithm with visualization

import matplotlib.pyplot as plt

def bresenhamLine(x1, y1, x2, y2):
    dx = x2 - x1
    dy = y2 - y1
    sx = 1 if dx >= 0 else -1
    sy = 1 if dy >= 0 else -1
    dx = abs(dx)
    dy = abs(dy)
    swap = False

    if dy > dx:
        swap = True
        dx, dy = dy, dx

    p = 2 * dy - dx
    x, y = x1, y1

    # List simpan koordinat pixel
    pixels = [(x, y)]

    for i in range(dx):
        if p >= 0:
            if swap:
                x += sx
            else:
                y += sy
            p -= 2 * dx

        if swap:
            y += sy
        else:
            x += sx
        p += 2 * dy

        # Menambahkan koordinat pixel ke list
        pixels.append((x, y))

    return pixels

# Membuat gambar menggunakan Matplotlib dengan label titik
def plot_line(pixels):
    x_values, y_values = zip(*pixels)
    plt.plot(x_values, y_values, marker='o', label='Pixel Coordinates')

    # Menambahkan label pada setiap titik
    for (x, y) in pixels:
        plt.text(x, y, f'({x}, {y})', fontsize=8, ha='right', va='bottom')

    plt.title('Bresenham Line Algorithm')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.grid(True)
    plt.legend()
    plt.show()

# Contoh pemanggilan
line_pixels = bresenhamLine(12, 10, -10, -15)
plot_line(line_pixels)
```

```python
# Python implementation of the Midpoint Circle Algorithm with visualization

import matplotlib.pyplot as plt

def midpoint_circle_algorithm(x0, y0, radius):
    x = radius
    y = 0
    radius_error = 1 - x

    points = []

    while x >= y:
        points.extend([
            (x + x0, y + y0), (y + x0, x + y0), (-x + x0, y + y0), (-y + x0, x + y0),
            (-x + x0, -y + y0), (-y + x0, -x + y0), (x + x0, -y + y0), (y + x0, -x + y0)
        ])

        y += 1

        if radius_error < 0:
            radius_error += 2 * y + 1
        else:
            x -= 1
            radius_error += 2 * (y - x + 1)

    return points

# Example usage of the function with center (4, 6) and radius 16
circle_points = midpoint_circle_algorithm(4, 6, 16)

# Plotting the result
x_coords, y_coords = zip(*circle_points)
plt.figure(figsize=(8, 8))
plt.scatter(x_coords, y_coords, color='blue')
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Midpoint Circle Algorithm Visualization")
plt.xlabel("X Coordinate")
plt.ylabel("Y Coordinate")
plt.grid(True)
plt.show()
```
